```python
# --- Clean ESG dataset ---

# 1. Drop series code (useless)
esg_clean = esg_df.drop(columns=["Series Code"])

# 2. Reshape from wide to long format
year_cols = [col for col in esg_clean.columns if "[" in col]  # columns like '2000 [YR2000]'

esg_long = esg_clean.melt(
    id_vars=["Country Name", "Country Code", "Series Name"],
    value_vars=year_cols,
    var_name="Year",
    value_name="Value"
)

# 3. Clean Year column: keep only the 4 digits
esg_long["Year"] = esg_long["Year"].str.slice(0, 4)

# 4. Convert Year to integer
esg_long["Year"] = esg_long["Year"].astype(int)

# Preview
esg_long.head()
```

```python
# --- Clean GDP/Inflation/FDI dataset ---

# 1. Drop Series Code (useless)
gdp_clean = gdp_df.drop(columns=["Series Code"])

# 2. Reshape from wide to long format
year_cols_gdp = [col for col in gdp_clean.columns if "[" in col]

gdp_long = gdp_clean.melt(
    id_vars=["Country Name", "Country Code", "Series Name"],
    value_vars=year_cols_gdp,
    var_name="Year",
    value_name="Value"
)

# 3. Clean Year column
gdp_long["Year"] = gdp_long["Year"].str.slice(0, 4)

# 4. Convert Year to integer
gdp_long["Year"] = gdp_long["Year"].astype(int)

# Preview
gdp_long.head()
```

```python
# --- Add Category to ESG dataset ---

def assign_category(series_name):
    series_name_lower = str(series_name).lower()  # changes every (if any) value in the series_name column from float to string + changes capital letters into normal letters
    
    # Environmental
    if "co2" in series_name_lower:
        return "Environmental"
    if "fossil" in series_name_lower:
        return "Environmental"
    if "renewable" in series_name_lower:
        return "Environmental"
    if "methane" in series_name_lower:
        return "Environmental"
    if "nitrous" in series_name_lower:
        return "Environmental"
    
    # Social
    if "unemployment" in series_name_lower:
        return "Social"
    if "gini" in series_name_lower:
        return "Social"
    if "rights" in series_name_lower:
        return "Social"

    # Governance
    if "corruption" in series_name_lower:
        return "Governance"
    if "political" in series_name_lower:
        return "Governance"

    # Economic
    if "gdp" in series_name_lower:
        return "Economic"
    if "expenditure" in series_name_lower:
        return "Economic"

    return "Other"

# Apply to ESG data
esg_long["Category"] = esg_long["Series Name"].apply(assign_category)

# Show each indicator in the dataset and its assigned category
esg_long[["Series Name", "Category"]].drop_duplicates().sort_values("Series Name")
```

```python
# ---Add Category to gdp-inflation-fdi dataset ---

def assign_economic_category(series_name):
    series_name_lower = str(series_name).lower() #same as above

    #Economic
    if "gdp" in series_name_lower:
        return "Economic"
    if "inflation" in series_name_lower:
        return "Economic"
    if "foreign direct investment" in series_name_lower:
        return "Economic"
    if "research" in series_name_lower or "r&d" in series_name_lower:
        return "Economic"

    return "Other"

gdp_long["Category"] = gdp_long["Series Name"].apply(assign_economic_category)

# Show each indicator in the dataset and its assigned category
gdp_long[["Series Name", "Category"]].drop_duplicates().sort_values("Series Name")
```

```python
# Drop rows where Series Name shows "Other" (NaN)
esg_long = esg_long.dropna(subset=["Series Name"])
gdp_long = gdp_long.dropna(subset=["Series Name"])
```

```python
# Show each indicator in the dataset and its assigned category
esg_long[["Series Name", "Category"]].drop_duplicates().sort_values("Series Name")
```

```python
# Show each indicator in the dataset and its assigned category
gdp_long[["Series Name", "Category"]].drop_duplicates().sort_values("Series Name")
```

```python
# Add a simple source label
esg_long["Source"] = "ESG"
gdp_long["Source"] = "Economic"

# Make column order consistent
common_cols = ["Country Name", "Country Code", "Series Name", "Year", "Value", "Category", "Source"]

esg_long = esg_long[common_cols]
gdp_long = gdp_long[common_cols]

# Combine into a single long dataframe
all_long = pd.concat([esg_long, gdp_long], ignore_index=True)

all_long.head(15)
```

```python
# sort the data by country code
all_long = all_long.sort_values(["Country Code", "Year", "Category"]).reset_index(drop=True)
all_long.head(20)
```

```python
# Prepare classification table: drop "lending category" and and rename columns to match the other (merged) dataset
class_clean = class_df.rename(
    columns={
        "Economy": "Country Name",
        "Code": "Country Code",
        "Income group": "Income Group",
        "Lending category": "Lending Category"
    }
)[["Country Code", "Country Name", "Region", "Income Group"]]

class_clean.head(15)
```

```python
panel_long = all_long.merge(
    class_clean,
    on="Country Code",
    how="left"
)

if "Country Name_x" in panel_long.columns:
    panel_long = panel_long.rename(columns={"Country Name_x": "Country Name"})
    panel_long = panel_long.drop(columns=["Country Name_y"])

panel_long.head(20)
```
